```python
#Logistic回归梯度上升优化算法
def loadDataSet():
    dataSet = [];labels = []
    
    
    fr  = open('E:/谢远东/机器学习/机器学习实践/机器学习实战源代码/machinelearninginaction/Ch05/testSet.txt')
    for line in fr.readlines():
        intArr = line.strip().split('\t')
        dataSet.append([1.0,float(intArr[0]),float(intArr[1])])
        labels.append(intArr[2])
    return dataSet,labels

def sigmoid(intX):
    return 1/(1+exp(intX))


import numpy as np
def gradAscent(dataMatIn,classLabels):
    dataMatrix = np.mat(dataMatIn)#转换成numpy数据
    labelMat = np.mat(classLabels.transpose())
    
    m,n = shape(dataMatrix)
    weights = ones(n,1)#定义第一次的权重，这里设置为单位向量
    maxRound = 500#迭代次数
    alpha = 0.01#α步长数
    for i in range(maxRound):
        h = sigmoid(dataMatrix*weights)
        error = (labelMat - h)
        weights = weights + alpha*dataMatrix.transpose()*error
    return weights
```
